**Вопрос 6**: На какой идее базируются алгоритмы аггломеративной кластеризации? Напишите формулу Ланса-Вильма

Идея: у нас есть какие-то точки в пространстве. Нужно их кластеризовать. Считаем каждую точку кластером. Точки, близко расположенные друг к другу, объединяем в один кластер. Повторяем. Получаем дерево. В итоге получим дерево, из которого можно выбрать кластеризацию с нужным максимально возможным расстоянием между элементами одного кластера.

Формула Ланса-Вильма:

Нужна для того, чтоб определить расстояние между кластерами W и S - $R(W,S)$, где $W = U \cup V$, зная R(U,S), R(V,S), R(U,V).

### $R(W,S) = {\alpha}_{U} \cdot R(U,S) + {\alpha}_{V} \cdot R(V,S) + {\beta}  \cdot R(U,V) + \gamma \cdot | R(U,S) - R(V,S) |$, где ${\alpha}_{U}, {\alpha}_{V}, \beta, \gamma$

**Вопрос 7**: Какие два шага выделяют в алгоритме кластеризации k-means?

1.Аналог E-шага

Векторы переразбиваются на кластеры. Каждый вектор $x_i$ относится к тому кластеру, к центру масс ${\mu}_{yj}$ которого он ближе находится - $y_i$.

### $y_i := {argmin}_{y \in Y} \rho(x_i, {\mu}_y), i = 1, \cdots l$


2.Аналог M-шага

Перевычисляется центр масс ${\mu}_{yj}$ для каждого кластера на основе векторов, отнесенных к нему.

### ${\mu}_{yj} := \frac {{\sum}_{i=1}^{l} [y_i = y]   f_j (x_i)} {{\sum}_{i=1}^{l} [y_i = y] }, y \in Y, j = 1, \cdots n$


**Вопрос 8**: В чём отличия (основные упрощения) k-means от EM-алгоритма кластеризации?

1.На E-шаге у EM осуществляется мягкая кластеризация, используем $g_{iy} = P{y_i = y}$, у k-means жёсткая кластеризация $g_{iy} = [y_i = y]$.


2.EM позволяет настраивать форму кластеров, k-means задаёт форму кластера метрикой $\rho$

** Вопрос 9 **Какой принцип работы графовых алгоритмов кластеризации?

Выборка представляется в виде графа. Вершины графа - объекты $x_i$, ребра графа - $\rho(x_i, x_j) \le R$, где $R$ задается.

Существуют разные графовые алгоритмы.

Например,

1.Алгоритм выделения главных компонент. Идея: удаляем ребра, расстояние на которых больше, чем R. Считаем число связанных компонент. Изменяем $R$,в зависимости от того, хотим мы увеличить число связанности или уменьшить, чтоб получить нужное количество кластеров.

2.Кратчайший незамкнутый путь. Находим пару вершин сребром, задающим наименьшим расстоянием,соединить их ребром. Покаввыборке остаются изолированные точки, находим ближайшую неизолированную точку к изолированной, соединяем их. Удаляем $K-1$ самых длинных ребер. $К$ - число кластеров, которое мы хотим.

```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F,np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[:i + 1]) / dTot for i in range(n)])
    #переход от элемента к следующему
    m = np.min([i if (var_i[i] > varRetained) else n-1 for i in range(n)]) + 1
    #вычислите необходимое число главных компонент
    print '%.3f variance retained in %d dimensions' % (var_i[m - 1], m )

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:,range(0,m)]
    G = np.dot(U_reduced,np.diag(d[:m]))

    return G, U_reduced
```

```python
from sklearn.datasets import load_digits
X = load_digits()
print(X.data.shape)
print(X.images.shape)

import matplotlib.pyplot as plt 
plt.gray() 
plt.matshow(X.images[0]) 
plt.show() 
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.cluster import adjusted_mutual_info_score, adjusted_rand_score

labels_true = X.target
```

```python
#переберем eps

for n in range(10,30):
    db = DBSCAN(eps=n)
    labels = db.fit_predict(X.data)
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    print 'eps', n, 'num_clusters', n_clusters_, 'noise', len([el for el in labels if el==-1]), 'a_mutual_info_s', \
    adjusted_mutual_info_score(labels, labels_true), 'a_rand_s', adjusted_rand_score(labels,labels_true)

# лучший результат по обоим метрикам
# eps 21 num_clusters 22 noise 321 a_mutual_info_s 0.725188049949 a_rand_s 0.613922600616
```

```python
#переберем min_samples

for n in range(2,10):
    db = DBSCAN(eps=21, min_samples = n)
    labels = db.fit_predict(X.data)
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    print 'min_sample', n, 'num_clusters', n_clusters_, 'noise', len([el for el in labels if el==-1]), 'a_mutual_info_s', \
    adjusted_mutual_info_score(labels, labels_true), 'a_rand_s', adjusted_rand_score(labels,labels_true)
    
#лучший min_samples = 4 для данного eps (были перебраны и другие)
```

```python
import scipy
from scipy import misc
import matplotlib.pyplot as plt 

image = misc.imread('two-lorikeet-birds-2293918.jpg')
plt.figure(figsize=(10,16))
plt.imshow(image)
plt.show()
print image.shape
X = image.reshape(-1,3)
print X.shape
```

```python
#-- KMeans

from sklearn.cluster import KMeans

num_clusters = [2, 3, 10, 20]
images = []

for el in num_clusters:
    X_data = np.array(X)
    print 'num_clusters =', el
    kmeans = KMeans(n_clusters=el)
    kmeans.fit_predict(X_data)
    print 'kmeans.cluster_centers'
    print kmeans.cluster_centers_, '\n'
    #print kmeans.labels_[:1000]
    for i in range(len(kmeans.labels_)):
        X_data[i] = (kmeans.cluster_centers_[kmeans.labels_[i]])
    images.append(X_data.reshape(275,400,3))
```

```python
n, m = 2, 2
plt.figure(figsize=(16,10))
for i in range(n):
    for j in range(m):
        k = i * m + j + 1
        plt.subplot(n, m, k)
        plt.imshow(images[k-1])
        plt.title('num_clusters = '+str(num_clusters[k-1]))
        plt.legend('best')
        plt.axis('off')
plt.show()
```

```python
#Для различных кластеризаций оцените и сравните потери от уменьшения цветов при помощи метрики SSIM

from skimage.measure import compare_ssim

print 'Kmeans \n'

for i,el in enumerate(num_clusters):
    print 'num_clusters = ' +str(el)+'; SSIM = '+str(compare_ssim(image, images[i],win_size=3))
```

```python
#-- MeanShift

from sklearn.cluster import estimate_bandwidth
from sklearn.cluster import MeanShift
```

```python
images = []
num_clusters = []
quantiles = [0.005, 0.02, 0.05, 0.08, 0.1, 0.2]
for i in quantiles:
    X_data = np.array(X)
    print 'quantile =', i
    bandwidth = estimate_bandwidth(X, quantile=i, n_samples=1000)
    ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    preds = ms.fit_predict(X_data)
    print 'num_clusters =', len(set(preds)), '\n'
    num_clusters.append(len(set(preds)))
    for i in range(len(ms.labels_)):
        X_data[i] = (ms.cluster_centers_[ms.labels_[i]])
    images.append(X_data.reshape(275,400,3))
```

```python
n, m = 2, 3
plt.figure(figsize=(16,10))
for i in range(n):
    for j in range(m):
        k = i * m + j + 1
        plt.subplot(n, m, k)
        plt.imshow(images[k-1])
        plt.title('quantiles = '+str(quantiles[k-1])+', num_clusters = '+str(num_clusters[k-1]))
        plt.legend('best')
        plt.axis('off')
plt.show()
```

```python
#Для различных кластеризаций оцените и сравните потери от уменьшения цветов при помощи метрики SSIM

from skimage.measure import compare_ssim

print 'MeanShift \n'

for i,el in enumerate(quantiles):
    print 'quantiles = ' +str(el)+'; SSIM = '+str(compare_ssim(image, images[i],win_size=3))
```

Видно,что при достаточно малом quantity (quantiles = 0.02; **SSIM = 0.834826977351**) **MeanShift** сравнима по качеству  с **Kmeans**(num_clusters = 20; **SSIM = 0.842976561352**) при достаточно большом количестве кластеров.

При уменьшении quantity в MeanShift SSIM увеличивается

При увеличении num_clusters в KMeans SSIM увеличивается

```python
#-- AgglomerativeClustering

from sklearn.cluster.hierarchical import AgglomerativeClustering
import numpy as np
import scipy as sc
from sklearn.neighbors import kneighbors_graph

num_clusters = [2, 3, 10, 20]
images = []

for el in num_clusters:
    X_data = np.array(X)
    print 'num_clusters =', el
    connectivity = kneighbors_graph(X_data, 30)
    model = AgglomerativeClustering(connectivity=connectivity, n_clusters=el)
    preds = model.fit_predict(X_data)
    unic_labels = list(sc.unique(preds))
    cluster_centers = []
    for label in unic_labels:
        cluster_centers.append(np.mean([X_data[i] for i in range(len(X_data)) if preds[i]==label], axis=0))
    new_image = []
    for i in range(len(X_data)):
        new_image.append(cluster_centers[unic_labels.index(preds[i])])
    new_image = np.array(new_image)
    images.append(new_image.reshape(275,400,3))
```
